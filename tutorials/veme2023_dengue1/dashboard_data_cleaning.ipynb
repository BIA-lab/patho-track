## Running Genome Detective
- Downloading all the fasta files for Africa from Genbank
- Running through the dengue typing tool on GD

```python
# downloading fasta files from NCBI 

from Bio import Entrez

Entrez.email = "your_email"

genbank_ids = genbank  

db = "nucleotide"  
rettype = "fasta"  
retmode = "text"  

sequences = []  

for genbank_id in genbank_ids:
    handle = Entrez.efetch(db=db, id=genbank_id, rettype=rettype, retmode=retmode)
    sequence = handle.read()  
    handle.close()
    sequences.append(sequence)
    
output_file = "dengue.fasta"  

with open(output_file, "w") as file:
    file.write("".join(sequences))

print(f"Sequences downloaded and saved to {output_file}")
```

```python
# changing the names back to the original GenBank accession IDs
# defining function to change this

def fix_name(row):
    name = row[0]
    
    if "." in name:
        return name.split(".")[0]
    
    else:
        return name
```

```python
# executing function

GD_data["name"] = GD_data.apply(fix_name, axis = "columns")
GD_data.head()
```

## Genomic Dash Framework metadata template

To fit on Genomic Dash Framework your csv file should have at least the following columns: 
- category_1: This should be the pathogen of your dashboard. For this tutorial we are going to use Dengue 1 (DENV-1)
- category_2: The sub-categories you want to show, named as will be shown in the dashboard
- date: date formatted
- region: Continent of the data
- country
- province: optional

For this tutorial we are only interested on Dengue virus type 1. So, let's filter the dataset

```python
dengue_1 = final_data[final_data["category_1"] == "dengue virus type 1"]
```

```python
dengue_1["category_2"].value_counts()
```

```python
# defining function to change genotype names
# row number needs to be the col = variants

def standardize_genotype(row):
    genotype = row["category_2"]
    
    if genotype == "Related to but not part of DENV-1 Genotype III":
        return "Unassigned"
    elif genotype == "DENV-1 Genotype I":
        return "Genotype I"
    elif genotype == "DENV-1 Genotype II":
        return "Genotype II"
    elif genotype == "DENV-1 Genotype III":
        return "Genotype III"
    elif genotype == "DENV-1 Genotype IV":
        return "Genotype IV"
    elif genotype == "DENV-1 Genotype V":
        return "Genotype V"
    elif genotype == "Could not assign":
        return "Unassigned"
    else:
        return genotype
```

```python
dengue_1["category_2"] = dengue_1.apply(standardize_genotype, axis = "columns")
```
